# Summary

This study explored how **demographic, lifestyle, and clinical factors** relate to metabolic markers (**HbA1c, fasting glucose, fasting insulin**) in U.S. young adults aged 18–39 using NHANES data.  

The analysis focused on identifying **modifiable risk factors** that could inform public health strategies for early-onset metabolic risk.

## Key Insights
- **Body weight (BMI)** and **physical activity** emerged as important factors linked to metabolic health.  
- Other factors (gender, age, sleep, alcohol, calorie, and sugar intake) showed no clear association.

## Public Health Implications
- Interventions should prioritize **weight management** and **regular exercise**.  
- Focus on **overall lifestyle patterns** rather than single nutrients or demographic traits.

## Data Limitations
- **Cross-sectional design** prevents causal inference.  
- **Self-reported measures** may be biased.  
- **Missing data and residual confounding** may affect observed associations.

## Overall Conclusion
Modifiable lifestyle factors are key determinants of metabolic health in young adults, and **longitudinal studies** are needed to establish causality.

## NHANES Analysis - Part 2: Data Filtering and Missing Data Analysis




Rationale for Part 2: Handling Missing Data for Analysis

This section addresses one of the most critical aspects of large-scale survey data analysis (missing data handling) to ensure the validity of comparisons across groups in the research question:

Key considerations and rationale:
- Preservation of Missing Data for Transparency
  
  The code avoids prematurely dropping missing values. Instead, it:

  - Preserves records with missing age values after filtering

  - Retains all original biomarker and demographic values, including NHANES-specific codes (e.g., 7777 = Refused, 9999 = Don't Know)

  - Enables researchers to trace the nature and extent of missingness without distorting raw data

- Detailed Missing Pattern Identification
  
  A dedicated function categorizes missingness into:

  - Standard missing (NaN)

  - Special codes (e.g., refused/don’t know)
    
    This distinction allows the researcher to understand why data is missing, not just how much is missing — which is essential when assessing whether data is missing at random (MAR), completely at random (MCAR), or not at random (MNAR).

- Indicator Variable Creation
  
  The code automatically generates binary indicator variables for each analysis variable, flagging:

  - Any missing value (_missing)

  - Specific types: refused (_refused), don’t know (_dont_know)
  
  These indicators can later be:

  - Used as covariates in regression models (e.g., to assess bias)

  - Used in sensitivity analyses to test how excluding or including missing categories impacts the results

- Assessment of Valid Data Availability
  
  For each key variable (e.g., glucose, HbA1c, insulin), the code quantifies the number and percentage of valid cases (excluding special codes). This helps determine:

  - Whether there's enough usable data for statistical comparison

  - Which variables may need imputation or should be interpreted cautiously

- Optional Creation of Analysis-Ready Variables
  
  Clean versions of variables are created (e.g., LBXGLU_analysis), replacing only special missing codes with NaN, while retaining the original variables. This dual structure supports:

  - Imputation or modeling with clean data

  - Auditable, transparent reporting with original codes preserved

```python
# NHANES Analysis - Part 2: Data Filtering and Missing Data Analysis
# Modified to preserve missing values for analysis consideration

def apply_age_filter(data, min_age=18, max_age=39):
    """Apply age filtering to the dataset"""
    print(f"\n{'='*60}")
    print("APPLYING AGE FILTER")
    print(f"{'='*60}")

    # Check if age variable exists
    if 'RIDAGEYR' not in data.columns:
        print("Age variable (RIDAGEYR) not found in dataset")
        return data

    print(f"Age filter: {min_age}-{max_age} years")
    print(f"Sample size before age filter: {len(data):,}")

    # Apply age filter - only filter on valid age values, preserve missing ages for analysis
    age_valid_mask = data['RIDAGEYR'].notna()
    age_range_mask = (data['RIDAGEYR'] >= min_age) & (data['RIDAGEYR'] <= max_age)

    # Keep participants with valid ages in range OR missing ages (for missing data analysis)
    filtered_data = data[age_range_mask | ~age_valid_mask].copy()

    valid_ages_in_range = data[age_valid_mask & age_range_mask]
    print(f"Sample size after age filter: {len(filtered_data):,}")
    print(f"  - Valid ages in range (18-39): {len(valid_ages_in_range):,}")
    print(f"  - Missing ages preserved: {(~age_valid_mask).sum():,}")
    print(f"Participants removed (valid ages outside range): {len(data) - len(filtered_data):,}")

    return filtered_data

def identify_missing_patterns(data):
    """Identify and categorize missing data patterns without replacing values"""
    print(f"\n{'='*60}")
    print("MISSING DATA PATTERN ANALYSIS")
    print(f"{'='*60}")

    missing_summary = []

    # Define NHANES special missing value codes and their meanings
    special_codes = {
        7777: "Refused",
        9999: "Don't know",
        77777: "Refused",
        99999: "Don't know",
        777777: "Refused",
        999999: "Don't know",
        7: "Refused (short)",
        9: "Don't know (short)",
        77: "Refused (medium)",
        99: "Don't know (medium)"
    }

    for col in data.columns:
        if col not in ['SEQN', 'CYCLE']:
            total_count = len(data)

            # Count standard missing (NaN/None)
            standard_missing = data[col].isna().sum()

            # Count special missing value codes
            special_missing_counts = {}
            total_special_missing = 0

            if data[col].dtype in ['int64', 'float64', 'object']:
                for code, meaning in special_codes.items():
                    count = (data[col] == code).sum()
                    if count > 0:
                        special_missing_counts[code] = {'count': count, 'meaning': meaning}
                        total_special_missing += count

            # Calculate valid observations
            total_missing = standard_missing + total_special_missing
            valid_count = total_count - total_missing
            missing_pct = (total_missing / total_count) * 100

            missing_summary.append({
                'Variable': col,
                'Total_Count': total_count,
                'Valid_Count': valid_count,
                'Standard_Missing': standard_missing,
                'Special_Missing_Total': total_special_missing,
                'Special_Missing_Details': special_missing_counts,
                'Total_Missing': total_missing,
                'Missing_Percentage': missing_pct,
                'Description': variable_info.get(col, {}).get('description', 'Unknown')
            })

    missing_df = pd.DataFrame(missing_summary)
    missing_df = missing_df.sort_values('Missing_Percentage', ascending=False)

    print("Missing Data Pattern Summary:")
    print("-" * 120)
    print(f"{'Variable':<12} {'Description':<25} {'Valid':<8} {'Std Missing':<11} {'Special Missing':<14} {'Total Missing':<13} {'Missing %':<10}")
    print("-" * 120)

    for _, row in missing_df.iterrows():
        print(f"{row['Variable']:<12} {row['Description'][:24]:<25} {row['Valid_Count']:<8} "
              f"{row['Standard_Missing']:<11} {row['Special_Missing_Total']:<14} "
              f"{row['Total_Missing']:<13} {row['Missing_Percentage']:<10.1f}")

    # Detailed special missing codes analysis
    print(f"\n{'='*60}")
    print("DETAILED SPECIAL MISSING CODES ANALYSIS")
    print(f"{'='*60}")

    for _, row in missing_df.iterrows():
        if row['Special_Missing_Details']:
            print(f"\n{row['Variable']} - {row['Description']}:")
            for code, details in row['Special_Missing_Details'].items():
                print(f"  Code {code} ({details['meaning']}): {details['count']} cases")

    return missing_df

def create_missing_indicators(data):
    """Create missing data indicator variables for analysis"""
    print(f"\n{'='*60}")
    print("CREATING MISSING DATA INDICATORS")
    print(f"{'='*60}")

    data_with_indicators = data.copy()

    # Define NHANES special missing value codes
    special_codes = [7, 9, 77, 99, 7777, 9999, 77777, 99999, 777777, 999999]

    analysis_vars = [var for var in variable_info.keys() if var in data.columns and var not in ['SEQN', 'CYCLE']]

    for var in analysis_vars:
        # Create indicator for any missing (standard + special codes)
        missing_mask = data[var].isna()

        # Add special missing codes to the mask
        if data[var].dtype in ['int64', 'float64']:
            for code in special_codes:
                missing_mask = missing_mask | (data[var] == code)

        # Create missing indicator variable
        indicator_name = f"{var}_missing"
        data_with_indicators[indicator_name] = missing_mask.astype(int)

        # Create specific indicators for different types of missing
        if data[var].dtype in ['int64', 'float64']:
            # Standard missing indicator
            data_with_indicators[f"{var}_standard_missing"] = data[var].isna().astype(int)

            # Refused indicator
            refused_mask = False
            for code in [7, 77, 7777, 77777, 777777]:
                refused_mask = refused_mask | (data[var] == code)
            data_with_indicators[f"{var}_refused"] = refused_mask.astype(int)

            # Don't know indicator
            dont_know_mask = False
            for code in [9, 99, 9999, 99999, 999999]:
                dont_know_mask = dont_know_mask | (data[var] == code)
            data_with_indicators[f"{var}_dont_know"] = dont_know_mask.astype(int)

    # Summary of indicators created
    indicator_vars = [col for col in data_with_indicators.columns if '_missing' in col or '_refused' in col or '_dont_know' in col]
    print(f"Created {len(indicator_vars)} missing data indicator variables:")

    for var in analysis_vars:
        indicators_for_var = [col for col in indicator_vars if col.startswith(var + '_')]
        if indicators_for_var:
            missing_total = data_with_indicators[f"{var}_missing"].sum()
            print(f"\n{var}:")
            print(f"  - {var}_missing: {missing_total} cases")
            if f"{var}_refused" in data_with_indicators.columns:
                refused_total = data_with_indicators[f"{var}_refused"].sum()
                print(f"  - {var}_refused: {refused_total} cases")
            if f"{var}_dont_know" in data_with_indicators.columns:
                dont_know_total = data_with_indicators[f"{var}_dont_know"].sum()
                print(f"  - {var}_dont_know: {dont_know_total} cases")

    return data_with_indicators

def analyze_valid_data_availability(data):
    """Analyze availability of valid (non-missing) data for analysis"""
    print(f"\n{'='*60}")
    print("VALID DATA AVAILABILITY FOR ANALYSIS")
    print(f"{'='*60}")

    # Define NHANES special missing value codes
    special_codes = [7, 9, 77, 99, 7777, 9999, 77777, 99999, 777777, 999999]

    analysis_vars = [var for var in variable_info.keys() if var in data.columns and var not in ['SEQN', 'CYCLE']]

    print(f"Analyzing valid data availability for {len(analysis_vars)} variables:")
    print("-" * 100)
    print(f"{'Variable':<12} {'Description':<30} {'Valid N':<10} {'Valid %':<10} {'Usable':<8}")
    print("-" * 100)

    availability_summary = []

    for var in analysis_vars:
        if var in data.columns:
            # Count valid observations (not NaN and not special missing codes)
            valid_mask = data[var].notna()

            if data[var].dtype in ['int64', 'float64']:
                for code in special_codes:
                    valid_mask = valid_mask & (data[var] != code)

            valid_count = valid_mask.sum()
            valid_pct = (valid_count / len(data)) * 100
            usable = "YES" if valid_count >= 100 else "NO"  # Minimum threshold for analysis

            description = variable_info.get(var, {}).get('description', 'Unknown')[:29]

            print(f"{var:<12} {description:<30} {valid_count:<10} {valid_pct:<10.1f} {usable:<8}")

            availability_summary.append({
                'variable': var,
                'valid_count': valid_count,
                'valid_percentage': valid_pct,
                'usable': usable
            })
        else:
            print(f"{var:<12} {'Variable not in dataset':<30} {0:<10} {0.0:<10.1f} {'NO':<8}")
            availability_summary.append({
                'variable': var,
                'valid_count': 0,
                'valid_percentage': 0.0,
                'usable': 'NO'
            })

    print("-" * 100)

    # Summary statistics
    usable_vars = [item for item in availability_summary if item['usable'] == 'YES']
    unusable_vars = [item for item in availability_summary if item['usable'] == 'NO']

    print(f"\nData Usability Summary:")
    print(f"  - Variables with sufficient data (≥100 valid cases): {len(usable_vars)}/{len(analysis_vars)}")
    print(f"  - Variables with insufficient data: {len(unusable_vars)}")

    if unusable_vars:
        print(f"  - Variables with insufficient data:")
        for item in unusable_vars:
            print(f"    * {item['variable']}: {item['valid_count']} valid cases")

    if usable_vars:
        avg_valid_pct = sum(item['valid_percentage'] for item in usable_vars) / len(usable_vars)
        min_valid = min(item['valid_count'] for item in usable_vars)
        max_valid = max(item['valid_count'] for item in usable_vars)

        print(f"\nUsable Variables Statistics:")
        print(f"  - Average valid data percentage: {avg_valid_pct:.1f}%")
        print(f"  - Minimum valid observations: {min_valid:,}")
        print(f"  - Maximum valid observations: {max_valid:,}")

    return availability_summary

def prepare_analysis_ready_data(data, preserve_original=True):
    """Prepare data for analysis while preserving original missing value information"""
    print(f"\n{'='*60}")
    print("PREPARING ANALYSIS-READY DATASET")
    print(f"{'='*60}")

    if preserve_original:
        # Keep original data intact and create analysis versions
        analysis_data = data.copy()

        # Add suffix to indicate these are analysis-ready versions
        special_codes = [7, 9, 77, 99, 7777, 9999, 77777, 99999, 777777, 999999]
        analysis_vars = [var for var in variable_info.keys() if var in data.columns and var not in ['SEQN', 'CYCLE']]

        for var in analysis_vars:
            if var in data.columns:
                # Create analysis version with special codes as NaN
                analysis_var_name = f"{var}_analysis"
                analysis_data[analysis_var_name] = data[var].copy()

                if data[var].dtype in ['int64', 'float64']:
                    for code in special_codes:
                        analysis_data.loc[analysis_data[var] == code, analysis_var_name] = np.nan

        print("Created analysis-ready variables with '_analysis' suffix")
        print("Original variables preserved with missing value codes intact")

        # Count analysis-ready variables
        analysis_ready_vars = [col for col in analysis_data.columns if col.endswith('_analysis')]
        print(f"Analysis-ready variables created: {len(analysis_ready_vars)}")

        return analysis_data

    else:
        print("Preserving original data structure - no modifications made")
        print("Use missing data indicators for analysis guidance")
        return data

# Apply filtering and analysis steps
if not nhanes_data.empty:
    # Step 1: Apply age filter (preserving missing ages for analysis)
    nhanes_filtered = apply_age_filter(nhanes_data, 18, 39)

    # Step 2: Analyze missing data patterns without modifying values
    missing_patterns = identify_missing_patterns(nhanes_filtered)

    # Step 3: Create missing data indicators
    nhanes_with_indicators = create_missing_indicators(nhanes_filtered)

    # Step 4: Analyze valid data availability
    valid_data_analysis = analyze_valid_data_availability(nhanes_with_indicators)

    # Step 5: Prepare analysis-ready dataset (optional)
    print(f"\n{'='*60}")
    print("DATASET PREPARATION OPTIONS")
    print(f"{'='*60}")
    print("Choose your approach:")
    print("1. Preserve all original values (recommended for missing data analysis)")
    print("2. Create analysis-ready versions alongside originals")

    # For this example, we'll preserve original values
    nhanes_final = nhanes_with_indicators.copy()

    # Final summary
    print(f"\n{'='*60}")
    print("FINAL DATASET SUMMARY")
    print(f"{'='*60}")
    print(f"Original sample size: {len(nhanes_data):,}")
    print(f"After age filtering: {len(nhanes_filtered):,}")
    print(f"Final dataset: {len(nhanes_final):,}")
    print(f"Total variables: {len(nhanes_final.columns)}")

    # Count different types of variables
    original_vars = [col for col in nhanes_final.columns if not any(suffix in col for suffix in ['_missing', '_refused', '_dont_know', '_analysis'])]
    indicator_vars = [col for col in nhanes_final.columns if any(suffix in col for suffix in ['_missing', '_refused', '_dont_know'])]

    print(f"  - Original variables: {len(original_vars)}")
    print(f"  - Missing data indicators: {len(indicator_vars)}")

    # Sample distribution by cycle
    if len(nhanes_final) > 0:
        print(f"\nSample distribution by cycle:")
        cycle_counts = nhanes_final['CYCLE'].value_counts().sort_index()
        for cycle, count in cycle_counts.items():
            pct = (count / len(nhanes_final)) * 100
            print(f"  {cycle}: {count:,} participants ({pct:.1f}%)")

    print(f"\n{'='*60}")
    print("ANALYSIS RECOMMENDATIONS")
    print(f"{'='*60}")
    print("1. Use missing data indicators to understand patterns")
    print("2. Consider multiple imputation for missing data")
    print("3. Perform sensitivity analyses excluding different missing categories")
    print("4. Report missing data patterns in your analysis")
    print("5. Original missing value codes preserved for transparency")

else:
    print("No data available for analysis")
```

### ✅ Rationale for Classification

- Thresholds are based on **established clinical guidelines** (ADA, WHO, CDC).
- Categorical groupings enable:
  - Cross-tabulations (e.g., % of obese females with high HbA1c)
  - Stratified comparison by **obesity, gender, and ethnicity**
  - Use in regression models and subgroup analyses
- Only **valid data** (excluding special/missing codes) are used for classification to ensure data quality and consistency.

---

These classifications prepare the dataset for statistical comparisons of glycemic biomarkers across demographic and health behavior subgroups, as required by the study objective.


```python
# NHANES Analysis - Part 3: Data Classification and Grouping (FIXED)

import pandas as pd
import numpy as np

def get_valid_data(data, variable):
    """Get valid data for a variable, excluding NHANES special missing codes"""
    if variable not in data.columns:
        return pd.Series(dtype='float64')

    # Define NHANES special missing value codes
    special_codes = [7, 9, 77, 99, 7777, 9999, 77777, 99999, 777777, 999999]

    # Start with non-NaN values
    valid_mask = data[variable].notna()

    # Exclude special missing codes
    if data[variable].dtype in ['int64', 'float64']:
        for code in special_codes:
            valid_mask = valid_mask & (data[variable] != code)

    return data.loc[valid_mask, variable]

def classify_fasting_glucose(data):
    """Classify fasting glucose levels"""
    print(f"\n{'='*60}")
    print("CLASSIFYING FASTING GLUCOSE (LBXGLU)")
    print(f"{'='*60}")

    if 'LBXGLU' not in data.columns:
        print("Fasting glucose variable (LBXGLU) not found")
        return data

    # Get valid data only
    valid_glucose = get_valid_data(data, 'LBXGLU')

    if len(valid_glucose) == 0:
        print("No valid glucose data available for classification")
        data['GLUCOSE_STATUS'] = np.nan
        return data

    # Initialize with NaN
    data['GLUCOSE_STATUS'] = np.nan

    # Apply classifications only to valid data
    valid_indices = valid_glucose.index
    data.loc[valid_indices, 'GLUCOSE_STATUS'] = pd.cut(valid_glucose,
                                                      bins=[-np.inf, 100, 126, np.inf],
                                                      labels=['Normal', 'Prediabetes', 'Diabetes'],
                                                      right=False)

    # Print distribution
    glucose_dist = data['GLUCOSE_STATUS'].value_counts().sort_index()
    total_valid = len(valid_glucose)
    total_missing = len(data) - total_valid

    print("Fasting Glucose Distribution:")
    print(f"  Normal (<100 mg/dL): {glucose_dist.get('Normal', 0):,}")
    print(f"  Prediabetes (100-125 mg/dL): {glucose_dist.get('Prediabetes', 0):,}")
    print(f"  Diabetes (≥126 mg/dL): {glucose_dist.get('Diabetes', 0):,}")
    print(f"  Valid classifications: {total_valid:,}")
    print(f"  Missing/Invalid data: {total_missing:,}")

    return data

def classify_hba1c(data):
    """Classify HbA1c levels"""
    print(f"\n{'='*60}")
    print("CLASSIFYING HbA1c (LBXGH)")
    print(f"{'='*60}")

    if 'LBXGH' not in data.columns:
        print("HbA1c variable (LBXGH) not found")
        return data

    # Get valid data only
    valid_hba1c = get_valid_data(data, 'LBXGH')

    if len(valid_hba1c) == 0:
        print("No valid HbA1c data available for classification")
        data['HBA1C_STATUS'] = np.nan
        return data

    # Initialize with NaN
    data['HBA1C_STATUS'] = np.nan

    # Apply classifications only to valid data
    valid_indices = valid_hba1c.index
    data.loc[valid_indices, 'HBA1C_STATUS'] = pd.cut(valid_hba1c,
                                                     bins=[-np.inf, 5.7, 6.5, np.inf],
                                                     labels=['Normal', 'Prediabetes', 'Diabetes'],
                                                     right=False)

    # Print distribution
    hba1c_dist = data['HBA1C_STATUS'].value_counts().sort_index()
    total_valid = len(valid_hba1c)
    total_missing = len(data) - total_valid

    print("HbA1c Distribution:")
    print(f"  Normal (<5.7%): {hba1c_dist.get('Normal', 0):,}")
    print(f"  Prediabetes (5.7-6.4%): {hba1c_dist.get('Prediabetes', 0):,}")
    print(f"  Diabetes (≥6.5%): {hba1c_dist.get('Diabetes', 0):,}")
    print(f"  Valid classifications: {total_valid:,}")
    print(f"  Missing/Invalid data: {total_missing:,}")

    return data

def classify_insulin(data):
    """Classify serum insulin levels"""
    print(f"\n{'='*60}")
    print("CLASSIFYING SERUM INSULIN (LBDINSI/INSFAST)")
    print(f"{'='*60}")

    # Check for insulin variables
    insulin_col = None
    for col in ['LBDINSI', 'INSFAST', 'LBXINS']:
        if col in data.columns:
            insulin_col = col
            break

    if insulin_col is None:
        print("Insulin variable (LBDINSI/INSFAST/LBXINS) not found")
        return data

    # Get valid data only
    valid_insulin = get_valid_data(data, insulin_col)

    if len(valid_insulin) == 0:
        print(f"No valid insulin data available for classification ({insulin_col})")
        data['INSULIN_STATUS'] = np.nan
        return data

    # Initialize with NaN
    data['INSULIN_STATUS'] = np.nan

    # Apply classifications only to valid data
    valid_indices = valid_insulin.index
    data.loc[valid_indices, 'INSULIN_STATUS'] = pd.cut(valid_insulin,
                                                      bins=[-np.inf, 15, 25, np.inf],
                                                      labels=['Normal', 'Elevated', 'High'],
                                                      right=False)

    # Print distribution
    insulin_dist = data['INSULIN_STATUS'].value_counts().sort_index()
    total_valid = len(valid_insulin)
    total_missing = len(data) - total_valid

    print(f"Serum Insulin Distribution ({insulin_col}):")
    print(f"  Normal (2-15 µU/mL): {insulin_dist.get('Normal', 0):,}")
    print(f"  Elevated (15-25 µU/mL): {insulin_dist.get('Elevated', 0):,}")
    print(f"  High (>25 µU/mL): {insulin_dist.get('High', 0):,}")
    print(f"  Valid classifications: {total_valid:,}")
    print(f"  Missing/Invalid data: {total_missing:,}")

    return data

def classify_bmi(data):
    """Classify BMI categories"""
    print(f"\n{'='*60}")
    print("CLASSIFYING BMI (BMXBMI)")
    print(f"{'='*60}")

    if 'BMXBMI' not in data.columns:
        print("BMI variable (BMXBMI) not found")
        return data

    # Get valid data only
    valid_bmi = get_valid_data(data, 'BMXBMI')

    if len(valid_bmi) == 0:
        print("No valid BMI data available for classification")
        data['BMI_CATEGORY'] = np.nan
        return data

    # Initialize with NaN
    data['BMI_CATEGORY'] = np.nan

    # Apply classifications only to valid data
    valid_indices = valid_bmi.index
    data.loc[valid_indices, 'BMI_CATEGORY'] = pd.cut(valid_bmi,
                                                     bins=[-np.inf, 18.5, 25, 30, np.inf],
                                                     labels=['Underweight', 'Normal', 'Overweight', 'Obese'],
                                                     right=False)

    # Print distribution
    bmi_dist = data['BMI_CATEGORY'].value_counts().sort_index()
    total_valid = len(valid_bmi)
    total_missing = len(data) - total_valid

    print("BMI Distribution:")
    print(f"  Underweight (<18.5): {bmi_dist.get('Underweight', 0):,}")
    print(f"  Normal (18.5-24.9): {bmi_dist.get('Normal', 0):,}")
    print(f"  Overweight (25-29.9): {bmi_dist.get('Overweight', 0):,}")
    print(f"  Obese (≥30): {bmi_dist.get('Obese', 0):,}")
    print(f"  Valid classifications: {total_valid:,}")
    print(f"  Missing/Invalid data: {total_missing:,}")

    return data

def classify_calorie_intake(data):
    """Classify total calorie intake"""
    print(f"\n{'='*60}")
    print("CLASSIFYING TOTAL CALORIE INTAKE (DR1TKCAL)")
    print(f"{'='*60}")

    if 'DR1TKCAL' not in data.columns:
        print("Total calorie intake variable (DR1TKCAL) not found")
        return data

    # Get valid data only
    valid_calories = get_valid_data(data, 'DR1TKCAL')

    if len(valid_calories) == 0:
        print("No valid calorie data available for classification")
        data['CALORIE_INTAKE'] = np.nan
        return data

    # Initialize with NaN
    data['CALORIE_INTAKE'] = np.nan

    # Apply classifications only to valid data
    valid_indices = valid_calories.index
    data.loc[valid_indices, 'CALORIE_INTAKE'] = pd.cut(valid_calories,
                                                       bins=[-np.inf, 1500, 2500, np.inf],
                                                       labels=['Low', 'Moderate', 'High'],
                                                       right=False)

    # Print distribution
    calorie_dist = data['CALORIE_INTAKE'].value_counts().sort_index()
    total_valid = len(valid_calories)
    total_missing = len(data) - total_valid

    print("Total Calorie Intake Distribution:")
    print(f"  Low (<1500 kcal): {calorie_dist.get('Low', 0):,}")
    print(f"  Moderate (1500-2499 kcal): {calorie_dist.get('Moderate', 0):,}")
    print(f"  High (≥2500 kcal): {calorie_dist.get('High', 0):,}")
    print(f"  Valid classifications: {total_valid:,}")
    print(f"  Missing/Invalid data: {total_missing:,}")

    return data

def classify_sugar_intake(data):
    """Classify added sugar intake"""
    print(f"\n{'='*60}")
    print("CLASSIFYING ADDED SUGAR INTAKE (DR1TSUGR)")
    print(f"{'='*60}")

    if 'DR1TSUGR' not in data.columns:
        print("Added sugar intake variable (DR1TSUGR) not found")
        return data

    # Get valid data only
    valid_sugar = get_valid_data(data, 'DR1TSUGR')

    if len(valid_sugar) == 0:
        print("No valid sugar data available for classification")
        data['SUGAR_INTAKE'] = np.nan
        return data

    # Initialize with NaN
    data['SUGAR_INTAKE'] = np.nan

    # Apply classifications only to valid data
    valid_indices = valid_sugar.index
    data.loc[valid_indices, 'SUGAR_INTAKE'] = pd.cut(valid_sugar,
                                                     bins=[-np.inf, 25, 50, np.inf],
                                                     labels=['Low', 'Moderate', 'High'],
                                                     right=False)

    # Print distribution
    sugar_dist = data['SUGAR_INTAKE'].value_counts().sort_index()
    total_valid = len(valid_sugar)
    total_missing = len(data) - total_valid

    print("Added Sugar Intake Distribution:")
    print(f"  Low (<25g): {sugar_dist.get('Low', 0):,}")
    print(f"  Moderate (25-49g): {sugar_dist.get('Moderate', 0):,}")
    print(f"  High (≥50g): {sugar_dist.get('High', 0):,}")
    print(f"  Valid classifications: {total_valid:,}")
    print(f"  Missing/Invalid data: {total_missing:,}")

    return data

def classify_alcohol_use(data):
    """Classify alcohol use patterns"""
    print(f"\n{'='*60}")
    print("CLASSIFYING ALCOHOL USE (ALQ101, ALQ130)")
    print(f"{'='*60}")

    # Check for alcohol variables
    if 'ALQ101' in data.columns:
        # Get valid data for ALQ101
        valid_alq101 = get_valid_data(data, 'ALQ101')

        # Initialize with NaN
        data['ALCOHOL_EVER'] = np.nan

        if len(valid_alq101) > 0:
            # Map only valid values
            alcohol_map = {1: 'Yes', 2: 'No'}
            data.loc[valid_alq101.index, 'ALCOHOL_EVER'] = valid_alq101.map(alcohol_map)

            alcohol_ever_dist = data['ALCOHOL_EVER'].value_counts()
            print("Ever had ≥12 drinks in lifetime:")
            for category, count in alcohol_ever_dist.items():
                if pd.notna(category):
                    print(f"  {category}: {count:,}")
            print(f"  Valid responses: {len(valid_alq101):,}")
            print(f"  Missing/Invalid: {len(data) - len(valid_alq101):,}")
        else:
            print("No valid ALQ101 data available")

    if 'ALQ130' in data.columns:
        # Get valid data for ALQ130
        valid_alq130 = get_valid_data(data, 'ALQ130')

        # Initialize with NaN
        data['ALCOHOL_FREQUENCY'] = np.nan

        if len(valid_alq130) > 0:
            # Map ALQ130 codes to frequency categories
            frequency_map = {
                1: 'Daily',
                2: 'Nearly Daily',
                3: '3-4 times/week',
                4: '2 times/week',
                5: 'Once/week',
                6: '2-3 times/month',
                7: 'Once/month',
                8: '7-11 times/year',
                9: '3-6 times/year',
                10: '1-2 times/year'
            }
            data.loc[valid_alq130.index, 'ALCOHOL_FREQUENCY'] = valid_alq130.map(frequency_map)

            alcohol_freq_dist = data['ALCOHOL_FREQUENCY'].value_counts()
            print("\nCurrent alcohol frequency:")
            for category, count in alcohol_freq_dist.items():
                if pd.notna(category):
                    print(f"  {category}: {count:,}")
            print(f"  Valid responses: {len(valid_alq130):,}")
            print(f"  Missing/Invalid: {len(data) - len(valid_alq130):,}")
        else:
            print("No valid ALQ130 data available")

    return data

def classify_sleep_hours(data):
    """Classify sleep hours"""
    print(f"\n{'='*60}")
    print("CLASSIFYING SLEEP HOURS (SLD010H)")
    print(f"{'='*60}")

    # Check for sleep variables
    sleep_col = None
    for col in ['SLD010H', 'SLD012']:
        if col in data.columns:
            sleep_col = col
            break

    if sleep_col is None:
        print("Sleep hours variable (SLD010H/SLD012) not found")
        return data

    # Get valid data only
    valid_sleep = get_valid_data(data, sleep_col)

    if len(valid_sleep) == 0:
        print(f"No valid sleep data available for classification ({sleep_col})")
        data['SLEEP_CATEGORY'] = np.nan
        return data

    # Initialize with NaN
    data['SLEEP_CATEGORY'] = np.nan

    # Apply classifications only to valid data
    valid_indices = valid_sleep.index
    data.loc[valid_indices, 'SLEEP_CATEGORY'] = pd.cut(valid_sleep,
                                                       bins=[-np.inf, 6, 9, np.inf],
                                                       labels=['Short', 'Adequate', 'Long'],
                                                       right=True)

    # Print distribution
    sleep_dist = data['SLEEP_CATEGORY'].value_counts().sort_index()
    total_valid = len(valid_sleep)
    total_missing = len(data) - total_valid

    print(f"Sleep Hours Distribution ({sleep_col}):")
    print(f"  Short (≤6 hours): {sleep_dist.get('Short', 0):,}")
    print(f"  Adequate (7-9 hours): {sleep_dist.get('Adequate', 0):,}")
    print(f"  Long (≥10 hours): {sleep_dist.get('Long', 0):,}")
    print(f"  Valid classifications: {total_valid:,}")
    print(f"  Missing/Invalid data: {total_missing:,}")

    return data

def classify_physical_activity(data):
    """Classify physical activity patterns"""
    print(f"\n{'='*60}")
    print("CLASSIFYING PHYSICAL ACTIVITY (PAQ605, PAQ620)")
    print(f"{'='*60}")

    if 'PAQ605' in data.columns:
        # Get valid data for PAQ605
        valid_paq605 = get_valid_data(data, 'PAQ605')

        # Initialize with NaN
        data['MODERATE_ACTIVITY'] = np.nan

        if len(valid_paq605) > 0:
            activity_map = {1: 'Yes', 2: 'No'}
            data.loc[valid_paq605.index, 'MODERATE_ACTIVITY'] = valid_paq605.map(activity_map)

            mod_activity_dist = data['MODERATE_ACTIVITY'].value_counts()
            print("Moderate Physical Activity:")
            for category, count in mod_activity_dist.items():
                if pd.notna(category):
                    print(f"  {category}: {count:,}")
            print(f"  Valid responses: {len(valid_paq605):,}")
        else:
            print("No valid PAQ605 data available")

    if 'PAQ620' in data.columns:
        # Get valid data for PAQ620
        valid_paq620 = get_valid_data(data, 'PAQ620')

        # Initialize with NaN
        data['VIGOROUS_ACTIVITY'] = np.nan

        if len(valid_paq620) > 0:
            activity_map = {1: 'Yes', 2: 'No'}
            data.loc[valid_paq620.index, 'VIGOROUS_ACTIVITY'] = valid_paq620.map(activity_map)

            vig_activity_dist = data['VIGOROUS_ACTIVITY'].value_counts()
            print("\nVigorous Physical Activity:")
            for category, count in vig_activity_dist.items():
                if pd.notna(category):
                    print(f"  {category}: {count:,}")
            print(f"  Valid responses: {len(valid_paq620):,}")
        else:
            print("No valid PAQ620 data available")

    return data

def classify_work_type(data):
    """Classify work type"""
    print(f"\n{'='*60}")
    print("CLASSIFYING WORK TYPE (PAQ650)")
    print(f"{'='*60}")

    if 'PAQ650' not in data.columns:
        print("Work type variable (PAQ650) not found")
        return data

    # Get valid data only
    valid_work = get_valid_data(data, 'PAQ650')

    # Initialize with NaN
    data['WORK_TYPE'] = np.nan

    if len(valid_work) > 0:
        work_map = {1: 'Sedentary', 2: 'Active'}
        data.loc[valid_work.index, 'WORK_TYPE'] = valid_work.map(work_map)

        work_dist = data['WORK_TYPE'].value_counts()
        print("Work Type Distribution:")
        for category, count in work_dist.items():
            if pd.notna(category):
                print(f"  {category}: {count:,}")
        print(f"  Valid responses: {len(valid_work):,}")
        print(f"  Missing/Invalid: {len(data) - len(valid_work):,}")
    else:
        print("No valid work type data available")

    return data

def classify_sedentary_behavior(data):
    """Classify sedentary behavior"""
    print(f"\n{'='*60}")
    print("CLASSIFYING SEDENTARY BEHAVIOR (PAD680)")
    print(f"{'='*60}")

    if 'PAD680' not in data.columns:
        print("Sedentary behavior variable (PAD680) not found")
        return data

    # Get valid data only
    valid_sedentary = get_valid_data(data, 'PAD680')

    if len(valid_sedentary) == 0:
        print("No valid sedentary behavior data available for classification")
        data['SEDENTARY_CATEGORY'] = np.nan
        return data

    # Initialize with NaN
    data['SEDENTARY_CATEGORY'] = np.nan

    # Apply classifications only to valid data
    valid_indices = valid_sedentary.index
    data.loc[valid_indices, 'SEDENTARY_CATEGORY'] = pd.cut(valid_sedentary,
                                                          bins=[-np.inf, 4, 7, np.inf],
                                                          labels=['Short', 'Moderate', 'Long'],
                                                          right=False)

    # Print distribution
    sedentary_dist = data['SEDENTARY_CATEGORY'].value_counts().sort_index()
    total_valid = len(valid_sedentary)
    total_missing = len(data) - total_valid

    print("Sedentary Behavior Distribution:")
    print(f"  Short (<4 hrs/day): {sedentary_dist.get('Short', 0):,}")
    print(f"  Moderate (4-7 hrs/day): {sedentary_dist.get('Moderate', 0):,}")
    print(f"  Long (>7 hrs/day): {sedentary_dist.get('Long', 0):,}")
    print(f"  Valid classifications: {total_valid:,}")
    print(f"  Missing/Invalid data: {total_missing:,}")

    return data
def classify_diabetes_diagnosis_age(data):
    """Classify age first diagnosed with diabetes (DID040)"""
    print(f"\n{'='*60}")
    print("CLASSIFYING AGE FIRST DIAGNOSED WITH DIABETES (DID040)")
    print(f"{'='*60}")

    if 'DID040' not in data.columns:
        print("DID040 variable not found in dataset")
        return data

    # Get valid data only
    valid_did040 = get_valid_data(data, 'DID040')

    if len(valid_did040) == 0:
        print("No valid age of diagnosis data available")
        data['AGE_DIAGNOSED_GROUP'] = np.nan
        return data

    # Initialize with NaN
    data['AGE_DIAGNOSED_GROUP'] = np.nan

    # Define bins and labels
    bins = [17, 19, 29, 39]
    labels = ['Diagnosed in teens', 'Diagnosed in 20s', 'Diagnosed in 30s']

    # Apply classification
    data.loc[valid_did040.index, 'AGE_DIAGNOSED_GROUP'] = pd.cut(valid_did040,
                                                                 bins=bins,
                                                                 labels=labels,
                                                                 right=True)

    # Print distribution
    diagnosis_dist = data['AGE_DIAGNOSED_GROUP'].value_counts().sort_index()
    for category, count in diagnosis_dist.items():
        print(f"  {category}: {count:,}")
    print(f"  Valid classifications: {len(valid_did040):,}")
    print(f"  Missing/Invalid data: {len(data) - len(valid_did040):,}")

    return data

def apply_all_classifications(data):
    """Apply all classification functions"""
    print(f"\n{'='*80}")
    print("APPLYING ALL CLASSIFICATIONS TO NHANES DATA")
    print(f"{'='*80}")

    # Apply all classifications
    data = classify_fasting_glucose(data)
    data = classify_hba1c(data)
    data = classify_insulin(data)
    data = classify_bmi(data)
    data = classify_calorie_intake(data)
    data = classify_sugar_intake(data)
    data = classify_alcohol_use(data)
    data = classify_sleep_hours(data)
    data = classify_physical_activity(data)
    data = classify_work_type(data)
    data = classify_sedentary_behavior(data)
    data = classify_diabetes_diagnosis_age(data)


    return data

def print_classification_summary(data):
    """Print summary of all classifications"""
    print(f"\n{'='*80}")
    print("CLASSIFICATION SUMMARY")
    print(f"{'='*80}")

    # List all new categorical variables created
    categorical_vars = [
        'GLUCOSE_STATUS', 'HBA1C_STATUS', 'INSULIN_STATUS', 'BMI_CATEGORY',
        'CALORIE_INTAKE', 'SUGAR_INTAKE', 'ALCOHOL_EVER', 'ALCOHOL_FREQUENCY',
        'SLEEP_CATEGORY', 'MODERATE_ACTIVITY', 'VIGOROUS_ACTIVITY',
        'WORK_TYPE', 'SEDENTARY_CATEGORY', 'AGE_DIAGNOSED_GROUP'
    ]

    print(f"Total participants in dataset: {len(data):,}")
    print(f"Number of new categorical variables created: {sum(1 for var in categorical_vars if var in data.columns)}")

    print("\nCategorical variables created:")
    for var in categorical_vars:
        if var in data.columns:
            valid_count = data[var].notna().sum()
            print(f"  ✓ {var} ({valid_count:,} valid classifications)")
        else:
            print(f"  ✗ {var} (variable not found in data)")

    return data

# Apply classifications to the dataset (using nhanes_final from Part 2)
if 'nhanes_final' in locals() and not nhanes_final.empty:
    print("Applying classifications to NHANES dataset...")
    nhanes_classified = apply_all_classifications(nhanes_final.copy())

    # Print classification summary
    nhanes_classified = print_classification_summary(nhanes_classified)

    print(f"\nDataset classified with {len(nhanes_classified):,} participants.")

    # Show final variable counts
    original_vars = [col for col in nhanes_classified.columns if not any(suffix in col for suffix in ['_missing', '_refused', '_dont_know', '_analysis', '_STATUS', '_CATEGORY', '_INTAKE', '_EVER', '_FREQUENCY', '_ACTIVITY', '_TYPE'])]
    indicator_vars = [col for col in nhanes_classified.columns if any(suffix in col for suffix in ['_missing', '_refused', '_dont_know'])]
    classification_vars = [col for col in nhanes_classified.columns if any(suffix in col for suffix in ['_STATUS', '_CATEGORY', '_INTAKE', '_EVER', '_FREQUENCY', '_ACTIVITY', '_TYPE'])]

    print(f"\nFinal variable counts:")
    print(f"  - Original variables: {len(original_vars)}")
    print(f"  - Missing data indicators: {len(indicator_vars)}")
    print(f"  - Classification variables: {len(classification_vars)}")
    print(f"  - Total variables: {len(nhanes_classified.columns)}")

else:
    print("Error: nhanes_final dataset not found. Please run Part 2 first to create the filtered dataset.")
```

```python
# NHANES Analysis - Part 5: Descriptive Statistics

def calculate_descriptive_stats(data):
    """Calculate comprehensive descriptive statistics"""
    print(f"\n{'='*60}")
    print("DESCRIPTIVE STATISTICS")
    print(f"{'='*60}")

    # Separate variables by type
    numeric_vars = []
    categorical_vars = []

    for col in data.columns:
        if col not in ['SEQN', 'CYCLE']:
            if data[col].dtype in ['int64', 'float64']:
                # Check if it's actually categorical (few unique values)
                unique_vals = data[col].nunique()
                if unique_vals <= 10 and data[col].min() >= 0:
                    categorical_vars.append(col)
                else:
                    numeric_vars.append(col)
            else:
                categorical_vars.append(col)

    return numeric_vars, categorical_vars

def describe_numeric_variables(data, numeric_vars):
    """Detailed descriptive statistics for numeric variables"""
    print(f"\n{'='*60}")
    print("NUMERIC VARIABLES SUMMARY")
    print(f"{'='*60}")

    if not numeric_vars:
        print("No numeric variables found for analysis")
        return

    # Create comprehensive statistics table
    stats_summary = []

    for var in numeric_vars:
        var_data = data[var].dropna()

        if len(var_data) > 0:
            stats = {
                'Variable': var,
                'N': len(var_data),
                'Mean': var_data.mean(),
                'Std': var_data.std(),
                'Min': var_data.min(),
                'Q1': var_data.quantile(0.25),
                'Median': var_data.median(),
                'Q3': var_data.quantile(0.75),
                'Max': var_data.max(),
                'IQR': var_data.quantile(0.75) - var_data.quantile(0.25),
                'Skewness': var_data.skew(),
                'Kurtosis': var_data.kurtosis(),
                'Description': variable_info.get(var, {}).get('description', 'Unknown')
            }
            stats_summary.append(stats)

    # Display results
    if stats_summary:
        print(f"{'Variable':<12} {'Description':<25} {'N':<8} {'Mean':<10} {'Std':<10} {'Min':<10} {'Max':<10}")
        print("-" * 95)

        for stat in stats_summary:
            print(f"{stat['Variable']:<12} {stat['Description'][:24]:<25} {stat['N']:<8} "
                  f"{stat['Mean']:<10.2f} {stat['Std']:<10.2f} {stat['Min']:<10.2f} {stat['Max']:<10.2f}")

        print(f"\n{'Variable':<12} {'Median':<10} {'Q1':<10} {'Q3':<10} {'IQR':<10} {'Skew':<8} {'Kurt':<8}")
        print("-" * 80)

        for stat in stats_summary:
            print(f"{stat['Variable']:<12} {stat['Median']:<10.2f} {stat['Q1']:<10.2f} {stat['Q3']:<10.2f} "
                  f"{stat['IQR']:<10.2f} {stat['Skewness']:<8.2f} {stat['Kurtosis']:<8.2f}")

    return stats_summary

def describe_categorical_variables(data, categorical_vars):
    """Detailed descriptive statistics for categorical variables"""
    print(f"\n{'='*60}")
    print("CATEGORICAL VARIABLES SUMMARY")
    print(f"{'='*60}")

    if not categorical_vars:
        print("No categorical variables found for analysis")
        return

    categorical_summary = []

    for var in categorical_vars:
        if var in data.columns:
            var_data = data[var].dropna()

            if len(var_data) > 0:
                value_counts = var_data.value_counts().sort_index()
                percentages = var_data.value_counts(normalize=True).sort_index() * 100

                print(f"\n{var} - {variable_info.get(var, {}).get('description', 'Unknown')}")
                print("-" * 50)
                print(f"{'Value':<10} {'Count':<10} {'Percentage':<12}")
                print("-" * 32)

                for value in value_counts.index:
                    count = value_counts[value]
                    pct = percentages[value]
                    print(f"{value:<10} {count:<10} {pct:<12.1f}%")

                # Summary statistics for categorical
                cat_stats = {
                    'Variable': var,
                    'N': len(var_data),
                    'Unique_Values': var_data.nunique(),
                    'Mode': var_data.mode().iloc[0] if len(var_data.mode()) > 0 else 'N/A',
                    'Mode_Frequency': value_counts.iloc[0] if len(value_counts) > 0 else 0,
                    'Mode_Percentage': percentages.iloc[0] if len(percentages) > 0 else 0,
                    'Description': variable_info.get(var, {}).get('description', 'Unknown')
                }
                categorical_summary.append(cat_stats)

    # Overall categorical summary
    if categorical_summary:
        print(f"\n{'='*50}")
        print("CATEGORICAL VARIABLES OVERVIEW")
        print(f"{'='*50}")
        print(f"{'Variable':<12} {'N':<8} {'Unique':<8} {'Mode':<8} {'Mode %':<8}")
        print("-" * 52)

        for cat in categorical_summary:
            print(f"{cat['Variable']:<12} {cat['N']:<8} {cat['Unique_Values']:<8} "
                  f"{cat['Mode']:<8} {cat['Mode_Percentage']:<8.1f}%")

    return categorical_summary

def analyze_demographics(data):
    """Analyze demographic characteristics"""
    print(f"\n{'='*60}")
    print("DEMOGRAPHIC ANALYSIS")
    print(f"{'='*60}")

    # Age analysis (if available)
    if 'RIDAGEYR' in data.columns:
        age_data = data['RIDAGEYR'].dropna()
        print(f"\nAge Distribution:")
        print(f"  Mean age: {age_data.mean():.1f} ± {age_data.std():.1f} years")
        print(f"  Median age: {age_data.median():.1f} years")
        print(f"  Age range: {age_data.min():.0f} - {age_data.max():.0f} years")

        # Age groups
        age_groups = pd.cut(age_data, bins=[17, 25, 30, 35, 40],
                           labels=['18-25', '26-30', '31-35', '36-39'])
        age_group_counts = age_groups.value_counts().sort_index()

        print(f"\n  Age Groups:")
        for group, count in age_group_counts.items():
            pct = (count / len(age_data)) * 100
            print(f"    {group}: {count:,} ({pct:.1f}%)")

    # Gender analysis (if available)
    gender_vars = ['RIAGENDR', 'GENDER', 'SEX']
    for var in gender_vars:
        if var in data.columns:
            gender_data = data[var].dropna()
            print(f"\nGender Distribution ({var}):")
            gender_counts = gender_data.value_counts().sort_index()
            for value, count in gender_counts.items():
                pct = (count / len(gender_data)) * 100
                gender_label = "Male" if value == 1 else "Female" if value == 2 else f"Value {value}"
                print(f"  {gender_label}: {count:,} ({pct:.1f}%)")
            break

    # Race/ethnicity analysis (if available)
    race_vars = ['RIDRETH1', 'RIDRETH3', 'RACE', 'ETHNICITY']
    for var in race_vars:
        if var in data.columns:
            race_data = data[var].dropna()
            print(f"\nRace/Ethnicity Distribution ({var}):")
            race_counts = race_data.value_counts().sort_index()
            for value, count in race_counts.items():
                pct = (count / len(race_data)) * 100
                print(f"  Category {value}: {count:,} ({pct:.1f}%)")
            break

    # Age first diagnosed with diabetes distribution (18–39 years only)
    if 'DID040' in data.columns:
        diagnosis_age_data = get_valid_data(data, 'DID040')

        # Filter to 18–39 years only
        filtered_diagnosis_age = diagnosis_age_data[(diagnosis_age_data >= 18) & (diagnosis_age_data <= 39)]

        # Bin into categories: teens, 20s, 30s
        if len(filtered_diagnosis_age) > 0:
            print(f"\nAge First Diagnosed with Diabetes Distribution (18–39 years):")
            diagnosis_bins = pd.cut(filtered_diagnosis_age,
                                    bins=[17, 19, 29, 39],
                                    labels=['Diagnosed in teens', 'Diagnosed in 20s', 'Diagnosed in 30s'],
                                    right=True)

            diagnosis_counts = diagnosis_bins.value_counts().sort_index()
            for category, count in diagnosis_counts.items():
                pct = (count / len(filtered_diagnosis_age)) * 100
                print(f"  {category}: {count:,} ({pct:.1f}%)")
        else:
            print("\nNo valid DID040 data found in age range 18–39.")
    else:
        print("\nDID040 variable not found in dataset.")


def detect_outliers(data, numeric_vars):
    """Detect outliers using IQR method"""
    print(f"\n{'='*60}")
    print("OUTLIER DETECTION (IQR Method)")
    print(f"{'='*60}")

    outlier_summary = []

    for var in numeric_vars:
        if var in data.columns:
            var_data = data[var].dropna()

            if len(var_data) > 0:
                Q1 = var_data.quantile(0.25)
                Q3 = var_data.quantile(0.75)
                IQR = Q3 - Q1

                lower_bound = Q1 - 1.5 * IQR
                upper_bound = Q3 + 1.5 * IQR

                outliers = var_data[(var_data < lower_bound) | (var_data > upper_bound)]
                outlier_count = len(outliers)
                outlier_pct = (outlier_count / len(var_data)) * 100

                outlier_info = {
                    'Variable': var,
                    'N': len(var_data),
                    'Outliers': outlier_count,
                    'Outlier_Percentage': outlier_pct,
                    'Lower_Bound': lower_bound,
                    'Upper_Bound': upper_bound,
                    'Min_Outlier': outliers.min() if outlier_count > 0 else None,
                    'Max_Outlier': outliers.max() if outlier_count > 0 else None
                }
                outlier_summary.append(outlier_info)

    # Display outlier summary
    if outlier_summary:
        print(f"{'Variable':<12} {'N':<8} {'Outliers':<10} {'Outlier %':<10} {'Lower Bound':<12} {'Upper Bound':<12}")
        print("-" * 74)

        for outlier in outlier_summary:
            print(f"{outlier['Variable']:<12} {outlier['N']:<8} {outlier['Outliers']:<10} "
                  f"{outlier['Outlier_Percentage']:<10.1f}% {outlier['Lower_Bound']:<12.2f} {outlier['Upper_Bound']:<12.2f}")

    return outlier_summary

def generate_data_summary(data):
    """Generate overall data summary"""
    print(f"\n{'='*60}")
    print("OVERALL DATA SUMMARY")
    print(f"{'='*60}")

    print(f"Dataset Shape: {data.shape[0]:,} rows × {data.shape[1]} columns")
    print(f"Memory Usage: {data.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

    # Data types summary
    dtype_counts = data.dtypes.value_counts()
    print(f"\nData Types:")
    for dtype, count in dtype_counts.items():
        print(f"  {dtype}: {count} variables")

    # Missing data overview
    total_cells = data.shape[0] * data.shape[1]
    missing_cells = data.isnull().sum().sum()
    missing_pct = (missing_cells / total_cells) * 100

    print(f"\nMissing Data Overview:")
    print(f"  Total cells: {total_cells:,}")
    print(f"  Missing cells: {missing_cells:,} ({missing_pct:.2f}%)")
    print(f"  Complete rows: {data.dropna().shape[0]:,}")

# Execute descriptive statistics analysis
if 'nhanes_final' in locals() and not nhanes_final.empty:
    # Identify variable types
    numeric_vars, categorical_vars = calculate_descriptive_stats(nhanes_final)

    print(f"\nVariable Classification:")
    print(f"  Numeric variables: {len(numeric_vars)} ({', '.join(numeric_vars)})")
    print(f"  Categorical variables: {len(categorical_vars)} ({', '.join(categorical_vars)})")

    # Generate comprehensive statistics
    numeric_stats = describe_numeric_variables(nhanes_final, numeric_vars)
    categorical_stats = describe_categorical_variables(nhanes_final, categorical_vars)

    # Demographic analysis
    analyze_demographics(nhanes_final)

    # Outlier detection
    outlier_results = detect_outliers(nhanes_final, numeric_vars)

    # Overall summary
    generate_data_summary(nhanes_final)

    print(f"\n{'='*60}")
    print("DESCRIPTIVE STATISTICS COMPLETE")
    print(f"{'='*60}")
    print(f"Analysis completed for {len(nhanes_final):,} participants")
    print(f"Ready for correlation analysis and visualization")

else:
    print("Error: nhanes_final dataset not found. Please run the data filtering steps first.")
```

## NHANES Analysis - Part 6: Data Visualization (Pure Visualization)

### Rationale

This code provides a comprehensive pipeline for visualizing NHANES data with a focus on pure exploratory visualization to understand data distributions, normality, categorical composition, group differences, and variable relationships. Here’s the breakdown and reasoning behind each visualization component:

---

### `create_numerical_histograms(data, numerical_vars)`
- **Purpose:**  
  Plot histograms for numeric variables to visualize their distributions.
- **Key points:**  
  - Uses Freedman-Diaconis rule for adaptive bin width to balance detail and smoothness.
  - Shows mean and median lines for quick comparison of central tendency and skewness.
  - Skips variables with fewer than 100 valid data points to ensure meaningful plots.
- **Why:**  
  Histograms reveal data shape, skew, modality, and potential data issues such as outliers or data entry errors.

---

### `create_qq_plots(data, numerical_vars)`
- **Purpose:**  
  Generate Q-Q plots for numeric variables to visually assess normality.
- **Key points:**  
  - Only includes variables with sufficient data (≥100 observations).
  - Q-Q plots compare data quantiles against theoretical normal distribution quantiles.
- **Why:**  
  Many statistical tests and models assume normality; Q-Q plots help assess this assumption visually.

---

### `create_categorical_bar_charts(data, categorical_vars)`
- **Purpose:**  
  Visualize the distribution of categorical variables using bar charts.
- **Key points:**  
  - Minimum threshold of 50 valid observations ensures reliable representation.
  - Bars show counts with overlaid percentages for intuitive understanding.
- **Why:**  
  Bar charts clarify the frequency and relative size of categories, revealing imbalances or dominant classes.

---

### `create_grouped_visualizations(data)`
- **Purpose:**  
  Show relationships between variables via grouped bar charts, boxplots, pie charts, and correlation heatmaps.
- **Key plots included:**  
  - Mean glucose by gender (grouped bar chart with error bars).  
  - HbA1c distribution by BMI category (boxplot).  
  - Pie chart of glucose status proportions.  
  - Correlation heatmap of key numeric variables.
- **Why:**  
  These plots illuminate how biomarkers differ across demographic or clinical groups and the interrelations between variables.

---

### `create_effect_heatmap(data)`
- **Purpose:**  
  Visualize the "effect sizes" or strength of associations between sets of independent variables (demographics, behavior) and dependent clinical outcomes (glucose, HbA1c, insulin status).
- **Method:**  
  - Uses coefficient of variation (CV) of row percentages in contingency tables as a proxy for effect size.  
  - Higher CV indicates greater variation and potential effect.
- **Why:**  
  Helps prioritize variables that show stronger influence on clinical statuses, guiding further detailed analyses.

---

### `create_comprehensive_dashboard(data)`
- **Purpose:**  
  Runs the full visualization suite in sequence:
  - Numerical histograms  
  - Q-Q plots  
  - Categorical bar charts  
  - Grouped visualizations  
  - Effect heatmap  
- **Why:**  
  Provides a one-stop, organized visual summary of the dataset for exploratory data analysis and reporting.

---

### Execution Logic
- The code checks for presence and non-emptiness of `nhanes_final` (preferred) or falls back to `complete_cases`.
- Runs the comprehensive dashboard and reports progress and completion status.
- If neither dataset exists, it prompts the user with instructions.

---

This visualization suite is designed to guide users through a robust exploratory phase, enhancing understanding of the NHANES dataset structure, quality, and key patterns before modeling or hypothesis testing.

```python
# NHANES Analysis - Part 6: Data Visualization (Pure Visualization)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Set visualization style
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 10

def create_numerical_histograms(data, numerical_vars, figsize=(15, 12)):
    """Create histograms for numerical variables with optimal binning"""
    print(f"\n{'='*60}")
    print("CREATING HISTOGRAMS FOR NUMERICAL VARIABLES")
    print(f"{'='*60}")

    # Filter variables that exist in data and have sufficient non-missing values
    available_vars = []
    for var in numerical_vars:
        if var in data.columns:
            valid_count = data[var].notna().sum()
            if valid_count >= 100:  # Minimum threshold for meaningful visualization
                available_vars.append(var)
            else:
                print(f"Skipping {var}: Only {valid_count} valid observations")

    if not available_vars:
        print("No numerical variables with sufficient data for visualization")
        return

    # Calculate grid dimensions
    n_vars = len(available_vars)
    n_cols = 3
    n_rows = (n_vars + n_cols - 1) // n_cols

    fig, axes = plt.subplots(n_rows, n_cols, figsize=figsize)
    if n_rows == 1:
        axes = axes.reshape(1, -1)
    axes = axes.flatten()

    for i, var in enumerate(available_vars):
        ax = axes[i]

        # Get valid data
        valid_data = data[var].dropna()

        # Determine optimal number of bins using Freedman-Diaconis rule
        if len(valid_data) > 0:
            q75, q25 = np.percentile(valid_data, [75, 25])
            iqr = q75 - q25
            if iqr > 0:
                bin_width = 2 * iqr / (len(valid_data) ** (1/3))
                optimal_bins = max(10, min(50, int((valid_data.max() - valid_data.min()) / bin_width)))
            else:
                optimal_bins = 30  # Default
        else:
            optimal_bins = 30

        # Create histogram
        ax.hist(valid_data, bins=optimal_bins, alpha=0.7, edgecolor='black', linewidth=0.5)

        # Add statistics to the plot
        mean_val = valid_data.mean()
        median_val = valid_data.median()
        std_val = valid_data.std()

        ax.axvline(mean_val, color='red', linestyle='--', alpha=0.8, label=f'Mean: {mean_val:.2f}')
        ax.axvline(median_val, color='blue', linestyle='--', alpha=0.8, label=f'Median: {median_val:.2f}')

        # Get variable description if available
        var_desc = var  # Simple label - can be enhanced with variable_info if available
        ax.set_title(f'{var}\n{var_desc}\n(n={len(valid_data):,}, std={std_val:.2f})', fontsize=9)
        ax.set_xlabel('Value')
        ax.set_ylabel('Frequency')
        ax.legend(fontsize=8)
        ax.grid(True, alpha=0.3)

    # Hide empty subplots
    for j in range(i + 1, len(axes)):
        axes[j].set_visible(False)

    plt.tight_layout()
    plt.suptitle('Distribution of Numerical Variables', fontsize=14, y=1.02)
    plt.show()

    return available_vars

def create_qq_plots(data, numerical_vars, figsize=(15, 12)):
    """Create Q-Q plots to visualize normality"""
    print(f"\n{'='*60}")
    print("CREATING Q-Q PLOTS FOR NORMALITY VISUALIZATION")
    print(f"{'='*60}")

    # Filter variables that exist in data
    available_vars = []
    for var in numerical_vars:
        if var in data.columns:
            valid_count = data[var].notna().sum()
            if valid_count >= 100:
                available_vars.append(var)

    if not available_vars:
        print("No numerical variables available for Q-Q plots")
        return

    # Calculate grid dimensions
    n_vars = len(available_vars)
    n_cols = 3
    n_rows = (n_vars + n_cols - 1) // n_cols

    fig, axes = plt.subplots(n_rows, n_cols, figsize=figsize)
    if n_rows == 1:
        axes = axes.reshape(1, -1)
    axes = axes.flatten()

    for i, var in enumerate(available_vars):
        ax = axes[i]

        # Get valid data
        valid_data = data[var].dropna()

        if len(valid_data) > 0:
            # Create Q-Q plot
            stats.probplot(valid_data, dist="norm", plot=ax)

            # Simple visual assessment of normality
            var_desc = var  # Simple label
            ax.set_title(f'{var}\n{var_desc}\n(n={len(valid_data):,})', fontsize=9)
            ax.grid(True, alpha=0.3)

    # Hide empty subplots
    for j in range(i + 1, len(axes)):
        axes[j].set_visible(False)

    plt.tight_layout()
    plt.suptitle('Q-Q Plots - Visual Assessment of Normality', fontsize=14, y=1.02)
    plt.show()

    return available_vars

def create_categorical_bar_charts(data, categorical_vars, base_height_per_row=5, n_cols=2):
    """Create labeled bar charts for categorical variables without squeezing"""
    print(f"\n{'='*60}")
    print("CREATING BAR CHARTS FOR CATEGORICAL VARIABLES")
    print(f"{'='*60}")

    # Filter variables that exist in data
    available_vars = []
    for var in categorical_vars:
        if var in data.columns:
            valid_count = data[var].notna().sum()
            if valid_count >= 50:  # Minimum threshold
                available_vars.append(var)
            else:
                print(f"Skipping {var}: Only {valid_count} valid observations")

    if not available_vars:
        print("No categorical variables with sufficient data for visualization")
        return

    # Calculate grid dimensions
    n_vars = len(available_vars)
    n_rows = (n_vars + n_cols - 1) // n_cols

    # Dynamically adjust figure height
    figsize = (16, base_height_per_row * n_rows)

    fig, axes = plt.subplots(n_rows, n_cols, figsize=figsize)
    if n_rows == 1:
        axes = axes.reshape(1, -1)
    axes = axes.flatten()

    for i, var in enumerate(available_vars):
        ax = axes[i]

        # Get value counts
        value_counts = data[var].value_counts().sort_index()

        if len(value_counts) > 0:
            # Create bar chart
            bars = ax.bar(range(len(value_counts)), value_counts.values, alpha=0.7)

            # Add value labels
            for j, (bar, count) in enumerate(zip(bars, value_counts.values)):
                percentage = (count / data[var].notna().sum()) * 100
                ax.text(bar.get_x() + bar.get_width()/2., bar.get_height() + bar.get_height()*0.01,
                        f'{count}\n({percentage:.1f}%)',
                        ha='center', va='bottom', fontsize=8)

            # Set labels
            ax.set_xticks(range(len(value_counts)))
            ax.set_xticklabels(value_counts.index, rotation=45, ha='right')

            var_desc = var
            total_valid = data[var].notna().sum()

            ax.set_title(f'{var}\n{var_desc}\nTotal valid observations: {total_valid:,}', fontsize=10)
            ax.set_ylabel('Count')
            ax.grid(True, alpha=0.3, axis='y')

    # Hide empty subplots
    for j in range(i + 1, len(axes)):
        axes[j].set_visible(False)

    # Adjust layout
    plt.tight_layout()
    plt.subplots_adjust(top=0.92)  # Prevent title from overlapping
    plt.suptitle('Distribution of Categorical Variables', fontsize=14, y=1.02)
    plt.show()

    return available_vars

def create_grouped_visualizations(data, figsize=(16, 12)):
    """Create grouped visualizations showing relationships between variables"""
    print(f"\n{'='*60}")
    print("CREATING GROUPED VISUALIZATIONS")
    print(f"{'='*60}")

    # Define key variables for analysis
    numerical_vars = ['LBXGLU', 'LBXGH', 'LBXIN']  # Glucose, HbA1c, Insulin
    categorical_factors = ['RIAGENDR', 'BMI_CATEGORY', 'GLUCOSE_STATUS', 'HBA1C_STATUS', 'AGE_DIAGNOSED_GROUP']

    # Filter available variables
    available_numerical = [var for var in numerical_vars if var in data.columns and data[var].notna().sum() >= 100]
    available_factors = [var for var in categorical_factors if var in data.columns and data[var].notna().sum() >= 50]

    if not available_numerical or not available_factors:
        print("Insufficient data for grouped visualizations")
        return

    fig, axes = plt.subplots(2, 2, figsize=figsize)
    axes = axes.flatten()

    plot_idx = 0

    # Plot 1: Grouped bar chart - Mean values by category
    if 'LBXGLU' in available_numerical and 'RIAGENDR' in available_factors:
        ax = axes[plot_idx]

        # Prepare data
        plot_data = data[['LBXGLU', 'RIAGENDR']].dropna()
        plot_data['Gender'] = plot_data['RIAGENDR'].map({1: 'Male', 2: 'Female'})

        # Create grouped bar chart
        gender_stats = plot_data.groupby('Gender')['LBXGLU'].agg(['mean', 'std', 'count']).reset_index()

        bars = ax.bar(gender_stats['Gender'], gender_stats['mean'],
                     yerr=gender_stats['std'], capsize=5, alpha=0.7)

        # Add value labels
        for bar, mean_val, count in zip(bars, gender_stats['mean'], gender_stats['count']):
            ax.text(bar.get_x() + bar.get_width()/2., bar.get_height() + bar.get_height()*0.01,
                   f'{mean_val:.1f}\n(n={count})', ha='center', va='bottom')

        ax.set_title('Mean Glucose Level by Gender')
        ax.set_ylabel('Glucose (mg/dL)')
        ax.grid(True, alpha=0.3, axis='y')
        plot_idx += 1

    # Plot 2: Box plot - Distribution by category
    if 'LBXGH' in available_numerical and 'BMI_CATEGORY' in available_factors:
        ax = axes[plot_idx]

        # Prepare data
        plot_data = data[['LBXGH', 'BMI_CATEGORY']].dropna()

        # Create box plot
        bmi_categories = sorted(plot_data['BMI_CATEGORY'].unique())
        bmi_data = [plot_data[plot_data['BMI_CATEGORY'] == cat]['LBXGH'].values
                   for cat in bmi_categories]

        box_plot = ax.boxplot(bmi_data, labels=bmi_categories, patch_artist=True)

        # Color the boxes
        colors = plt.cm.Set3(np.linspace(0, 1, len(box_plot['boxes'])))
        for patch, color in zip(box_plot['boxes'], colors):
            patch.set_facecolor(color)
            patch.set_alpha(0.7)

        ax.set_title('HbA1c Distribution by BMI Category')
        ax.set_ylabel('HbA1c (%)')
        ax.set_xlabel('BMI Category')
        ax.grid(True, alpha=0.3, axis='y')
        plt.setp(ax.get_xticklabels(), rotation=45, ha='right')
        plot_idx += 1

    # Plot 3: Pie chart for categorical distribution
    if 'GLUCOSE_STATUS' in available_factors:
        ax = axes[plot_idx]

        glucose_counts = data['GLUCOSE_STATUS'].value_counts()

        # Create pie chart
        wedges, texts, autotexts = ax.pie(glucose_counts.values, labels=glucose_counts.index,
                                         autopct='%1.1f%%', startangle=90)

        # Add count labels
        for i, (wedge, count) in enumerate(zip(wedges, glucose_counts.values)):
            angle = (wedge.theta2 + wedge.theta1) / 2
            x = wedge.r * 0.7 * np.cos(np.radians(angle))
            y = wedge.r * 0.7 * np.sin(np.radians(angle))
            ax.text(x, y, f'n={count}', ha='center', va='center', fontweight='bold')

        ax.set_title('Distribution of Glucose Status')
        plot_idx += 1

    # Plot 4: Correlation heatmap for numerical variables
    if len(available_numerical) >= 2:
        ax = axes[plot_idx]

        # Calculate correlation matrix
        corr_data = data[available_numerical].corr()

        # Create heatmap
        sns.heatmap(corr_data, annot=True, cmap='coolwarm', center=0,
                   square=True, ax=ax, cbar_kws={'shrink': 0.8})

        ax.set_title('Correlation Matrix: Key Variables')
        plot_idx += 1

    # Hide unused subplots
    for j in range(plot_idx, len(axes)):
        axes[j].set_visible(False)

    plt.tight_layout()
    plt.suptitle('Grouped Visualizations - Variable Relationships', fontsize=16, y=1.02)
    plt.show()

def create_effect_heatmap(data, figsize=(14, 10)):
    """Create heatmap showing effects of variables on dependent variables"""
    print(f"\n{'='*60}")
    print("CREATING EFFECT HEATMAP")
    print(f"{'='*60}")

    # Define independent and dependent variables
    independent_vars = ['RIAGENDR', 'BMI_CATEGORY', 'CALORIE_INTAKE', 'SUGAR_INTAKE',
                       'ALCOHOL_FREQUENCY', 'SLEEP_CATEGORY', 'MODERATE_ACTIVITY', 'VIGOROUS_ACTIVITY', 'AGE_DIAGNOSED_GROUP']
    dependent_vars = ['GLUCOSE_STATUS', 'HBA1C_STATUS', 'INSULIN_STATUS']

    # Filter available variables
    available_independent = [var for var in independent_vars if var in data.columns and data[var].notna().sum() >= 50]
    available_dependent = [var for var in dependent_vars if var in data.columns and data[var].notna().sum() >= 50]

    if len(available_independent) < 2 or len(available_dependent) < 1:
        print("Insufficient variables for effect heatmap")
        return

    # Calculate mean differences for visualization
    effect_matrix = pd.DataFrame(index=available_independent, columns=available_dependent)

    for indep_var in available_independent:
        for dep_var in available_dependent:
            # Create crosstab for visualization
            crosstab = pd.crosstab(data[indep_var], data[dep_var], dropna=False)

            if crosstab.size > 1:
                # Calculate normalized effect size for visualization
                # Using row percentages to show distribution patterns
                row_pct = crosstab.div(crosstab.sum(axis=1), axis=0)
                # Use coefficient of variation as a measure of effect
                cv_values = []
                for col in row_pct.columns:
                    if row_pct[col].std() > 0:
                        cv = row_pct[col].std() / row_pct[col].mean()
                        cv_values.append(cv)

                if cv_values:
                    effect_size = np.mean(cv_values)
                else:
                    effect_size = 0

                effect_matrix.loc[indep_var, dep_var] = effect_size
            else:
                effect_matrix.loc[indep_var, dep_var] = 0

    # Convert to numeric
    effect_matrix = effect_matrix.astype(float)

    # Create heatmap
    plt.figure(figsize=figsize)
    sns.heatmap(effect_matrix, annot=True, cmap='YlOrRd', fmt='.3f',
               vmin=0, square=True, cbar_kws={'label': "Effect Size"})

    plt.title('Effect Heatmap: Independent Variables vs Dependent Variables\n(Higher values indicate larger effects)')
    plt.xlabel('Dependent Variables')
    plt.ylabel('Independent Variables')
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

    return effect_matrix

def create_comprehensive_dashboard(data):
    """Create a comprehensive visualization dashboard"""
    print(f"\n{'='*80}")
    print("CREATING COMPREHENSIVE VISUALIZATION DASHBOARD")
    print(f"{'='*80}")

    # Define variable categories
    numerical_vars = ['RIDAGEYR', 'LBXGLU', 'LBXGH', 'LBXIN', 'BMXBMI',
                     'DR1TKCAL', 'DR1TSUGR', 'SLD010H', 'DID040']

    categorical_vars = ['RIAGENDR', 'GLUCOSE_STATUS', 'HBA1C_STATUS', 'INSULIN_STATUS',
                       'BMI_CATEGORY', 'CALORIE_INTAKE', 'SUGAR_INTAKE', 'ALCOHOL_EVER',
                       'ALCOHOL_FREQUENCY', 'SLEEP_CATEGORY', 'MODERATE_ACTIVITY',
                       'VIGOROUS_ACTIVITY', 'WORK_TYPE', 'SEDENTARY_CATEGORY', 'AGE_DIAGNOSED_GROUP']

    print(f"Starting visualization process for {len(data):,} participants...")

    # 1. Numerical variable histograms
    print("\n1. Creating histograms for numerical variables...")
    hist_vars = create_numerical_histograms(data, numerical_vars)

    # 2. Q-Q plots for normality visualization
    print("\n2. Creating Q-Q plots for normality visualization...")
    qq_vars = create_qq_plots(data, numerical_vars)

    # 3. Categorical variable bar charts
    print("\n3. Creating bar charts for categorical variables...")
    cat_vars = create_categorical_bar_charts(data, categorical_vars)

    # 4. Grouped visualizations
    print("\n4. Creating grouped visualizations...")
    create_grouped_visualizations(data)

    # 5. Effect heatmap
    print("\n5. Creating effect heatmap...")
    effect_results = create_effect_heatmap(data)

    print(f"\n{'='*80}")
    print("VISUALIZATION SUMMARY")
    print(f"{'='*80}")
    print(f"✅ Histograms created for {len(hist_vars) if hist_vars else 0} numerical variables")
    print(f"✅ Q-Q plots created for {len(qq_vars) if qq_vars else 0} variables")
    print(f"✅ Bar charts created for {len(cat_vars) if cat_vars else 0} categorical variables")
    print("✅ Grouped visualizations completed")
    print("✅ Effect heatmap completed")

    return {
        'numerical_vars_plotted': hist_vars,
        'qq_vars_plotted': qq_vars,
        'categorical_vars_plotted': cat_vars,
        'effect_matrix': effect_results
    }

# Execute visualization pipeline
if 'nhanes_final' in locals() and not nhanes_final.empty:
    print("Starting comprehensive data visualization...")

    # Use the complete dataset for visualizations
    viz_results = create_comprehensive_dashboard(nhanes_final)

    print(f"\n🎉 SUCCESS: Comprehensive visualizations completed!")
    print(f"📊 Dataset: {len(nhanes_final):,} participants")
    print(f"📈 All visualization types generated successfully")

elif 'complete_cases' in locals() and not complete_cases.empty:
    print("Using complete cases subset for visualization...")
    viz_results = create_comprehensive_dashboard(complete_cases)

    print(f"\n🎉 SUCCESS: Visualizations completed using complete cases!")
    print(f"📊 Dataset: {len(complete_cases):,} participants")

else:
    print("❌ ERROR: No suitable dataset found for visualization.")
    print("Please ensure your data is loaded in variables 'nhanes_final' or 'complete_cases'.")
    print("Example usage:")
    print("# Assuming your data is in 'df':")
    print("viz_results = create_comprehensive_dashboard(df)")
```

Function-by-Function Rationale

### 1. `load_merged_diq_with_age()`
- Merges DIQ (diabetes questionnaire) with DEMO (demographics) for each cycle.

- Ensures DIQ data can be filtered to the correct age group (18–39) and linked to trends.

### 2. `summarize_medians_and_diagnoses()`
- Computes **medians** of biomarkers and lifestyle indicators:
  - `LBXGLU` (fasting glucose)
  - `LBXGH` (HbA1c)
  - `LBDINSI` (insulin)
  - `BMXBMI`, `DR1TKCAL`, `SLD010H`, etc.
- Calculates **prevalence of high-risk labels** in classified variables:
  - e.g., “Diabetes” in `GLUCOSE_STATUS`
- Uses DIQ data to report:
  - % diagnosed with diabetes (`DIQ010`)
  - Median age of diagnosis (`DID040`)
  - % currently taking insulin (`DIQ050`)

### 3. `plot_trends()`
- Plots **biomarker medians** across cycles for:
  - `LBXGLU`, `LBXGH`, `LBDINSI`, `BMXBMI`
- Plots **prevalence trends** of high-risk groups across cycles:
  - Obesity, diabetes, insulin resistance, high calorie/sugar intake
- Visualizes **diagnosis and insulin use trends** using DIQ data

### Value to Research Objective

This summary and visualization code helps:
- **Describe** the 18–39 cohort across NHANES years
- **Highlight temporal changes** in metabolic risk
- **Quantify group prevalence** of high-risk traits
- Prepare the dataset for the **next analytical step**: group comparisons of glucose, HbA1c, and insulin across dependent variables



```python
# NHANES Analysis - Part 7: Summary and Trend Plots

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# NHANES DIQ file URLs (2005–2018)
diq_urls = {
    "2005-2006": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2005/DataFiles/DIQ_D.xpt",
    "2007-2008": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/DIQ_E.xpt",
    "2009-2010": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/DIQ_F.xpt",
    "2011-2012": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/DIQ_G.xpt",
    "2013-2014": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2013/DataFiles/DIQ_H.xpt",
    "2015-2016": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2015/DataFiles/DIQ_I.xpt",
    "2017-2018": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/DIQ_J.xpt"
}

def load_merged_diq_with_age():
    """Download and merge DIQ with DEMO for RIDAGEYR by cycle"""
    all_diq = []
    for cycle, diq_url in diq_urls.items():
        try:
            df_diq = pd.read_sas(diq_url)
            demo_url = diq_url.replace("DIQ", "DEMO")
            df_demo = pd.read_sas(demo_url)
            merged = pd.merge(df_diq, df_demo[['SEQN', 'RIDAGEYR']], on='SEQN', how='left')
            merged['CYCLE'] = cycle
            all_diq.append(merged)
            print(f"✅ Loaded {cycle} with age data")
        except Exception as e:
            print(f"❌ Failed to load {cycle}: {e}")
    return pd.concat(all_diq, ignore_index=True)

def summarize_medians_and_diagnoses(data, diq_data=None):
    print(f"\n{'='*70}")
    print("MEDIANS, HIGH-RISK GROUPS & DIABETES DIAGNOSES (Ages 18-39)")
    print(f"{'='*70}")

    if 'RIDAGEYR' in data.columns:
        data = data[(data['RIDAGEYR'] >= 18) & (data['RIDAGEYR'] <= 39)]
        print(f"Filtered dataset to age 18–39: {len(data):,} participants")

    continuous_vars = [
        'LBXGLU', 'LBXGH', 'LBDINSI', 'BMXBMI',
        'DR1TKCAL', 'DR1TSUGR', 'PAD680', 'SLD010H', 'RIDAGEYR', 'DID040'
    ]
    vars_present = [v for v in continuous_vars if v in data.columns]
    medians = data[vars_present].median().sort_values()

    print("\nMedian values (continuous variables):")
    for var, val in medians.items():
        print(f"  {var:<12}: {val:.2f}")

    classification_vars = {
        'GLUCOSE_STATUS': 'Diabetes',
        'HBA1C_STATUS': 'Diabetes',
        'INSULIN_STATUS': 'High',
        'BMI_CATEGORY': 'Obese',
        'SEDENTARY_CATEGORY': 'Long',
        'SLEEP_CATEGORY': 'Short',
        'SUGAR_INTAKE': 'High',
        'CALORIE_INTAKE': 'High',
        'ALCOHOL_FREQUENCY': 'High'
    }

    print("\nHigh-risk/abnormal category prevalence:")
    for var, high_risk_label in classification_vars.items():
        if var in data.columns:
            total = data[var].notna().sum()
            high = (data[var] == high_risk_label).sum()
            pct = (high / total) * 100 if total > 0 else float('nan')
            print(f"  {var:<20} ({high_risk_label}): {high:,}/{total:,} ({pct:.1f}%)")

    if 'BMI_CATEGORY' in data.columns:
        print("\nBMI Category Distribution:")
        for cat, count in data['BMI_CATEGORY'].value_counts().sort_index().items():
            pct = count / len(data) * 100
            print(f"  {cat:<12}: {count:,} ({pct:.1f}%)")

    if diq_data is not None and 'RIDAGEYR' in diq_data.columns:
        diq_data = diq_data[(diq_data['RIDAGEYR'] >= 18) & (diq_data['RIDAGEYR'] <= 39)]

        if 'DIQ010' in diq_data.columns:
            diq010_valid = diq_data['DIQ010'].dropna()
            diagnosed = diq010_valid[diq010_valid == 1]
            pct_diagnosed = (len(diagnosed) / len(diq010_valid)) * 100
            print(f"\nDiagnosed with diabetes (DIQ010): {len(diagnosed):,}/{len(diq010_valid):,} ({pct_diagnosed:.1f}%)")

        if 'DID040' in diq_data.columns:
            valid_ages = diq_data['DID040'].dropna()
            valid_ages = valid_ages[(valid_ages >= 18) & (valid_ages <= 39)]
            if len(valid_ages) > 0:
                print(f"\nAge at First Diabetes Diagnosis (DID040):")
                print(f"  Valid responses: {len(valid_ages):,}")
                print(f"  Median: {valid_ages.median():.1f} years")
                print(f"  Range: {valid_ages.min():.0f} – {valid_ages.max():.0f} years")

        if 'DIQ050' in diq_data.columns:
            diq050_valid = diq_data['DIQ050'].dropna()
            taking_insulin = diq050_valid[diq050_valid == 1]
            pct_insulin = (len(taking_insulin) / len(diq050_valid)) * 100
            print(f"\nCurrently taking insulin (DIQ050): {len(taking_insulin):,}/{len(diq050_valid):,} ({pct_insulin:.1f}%)")
    else:
        print("\nDIQ dataset not provided or missing RIDAGEYR")

    return medians

def plot_trends(data, diq_data=None):
    print("\n📊 Plotting trends across NHANES cycles...\n")

    if 'CYCLE' not in data.columns:
        print("No 'CYCLE' column in main dataset; cannot plot trends.")
        return

    cycle_order = sorted(data['CYCLE'].unique(), key=lambda x: int(x.split('-')[0]))
    data['CYCLE'] = pd.Categorical(data['CYCLE'], categories=cycle_order, ordered=True)
    data = data[(data['RIDAGEYR'] >= 18) & (data['RIDAGEYR'] <= 39)]

    # Plot median biomarkers
    median_vars = ['LBXGLU', 'LBXGH', 'LBDINSI', 'BMXBMI']
    plt.figure(figsize=(14, 8))
    for var in median_vars:
        if var in data.columns:
            med = data.groupby('CYCLE')[var].median()
            sns.lineplot(x=med.index, y=med.values, label=var)
    plt.title('Median Biomarker Trends (Age 18–39)')
    plt.xlabel('NHANES Cycle')
    plt.ylabel('Median Value')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Plot % high-risk categories
    risk_vars = {
        'GLUCOSE_STATUS': 'Diabetes',
        'HBA1C_STATUS': 'Diabetes',
        'INSULIN_STATUS': 'High',
        'BMI_CATEGORY': 'Obese',
        'SEDENTARY_CATEGORY': 'Long',
        'SLEEP_CATEGORY': 'Short',
        'SUGAR_INTAKE': 'High',
        'CALORIE_INTAKE': 'High',
        'ALCOHOL_FREQUENCY': 'High'
    }

    plt.figure(figsize=(16, 10))
    for var, label in risk_vars.items():
        if var in data.columns:
            pct = data.groupby('CYCLE').apply(
                lambda df: (df[var] == label).sum() / df[var].notna().sum() * 100 if df[var].notna().sum() > 0 else np.nan
            )
            sns.lineplot(x=pct.index, y=pct.values, label=var)
    plt.title('High-Risk Prevalence by NHANES Cycle (Age 18–39)')
    plt.ylabel('Percent (%)')
    plt.xlabel('NHANES Cycle')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Plot DIQ010 and DIQ050 trends if available
    if diq_data is not None and 'CYCLE' in diq_data.columns:
        diq_data = diq_data[(diq_data['RIDAGEYR'] >= 18) & (diq_data['RIDAGEYR'] <= 39)]
        diq_data['CYCLE'] = pd.Categorical(diq_data['CYCLE'], categories=cycle_order, ordered=True)

        if 'DIQ010' in diq_data.columns:
            diag = diq_data.groupby('CYCLE').apply(lambda df: (df['DIQ010'] == 1).sum() / df['DIQ010'].notna().sum() * 100)
            plt.figure(figsize=(10,6))
            sns.lineplot(x=diag.index, y=diag.values, marker='o')
            plt.title('Diagnosed Diabetes (%) by NHANES Cycle')
            plt.ylabel('Percent Diagnosed')
            plt.xlabel('NHANES Cycle')
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()

        if 'DIQ050' in diq_data.columns:
            insulin = diq_data.groupby('CYCLE').apply(lambda df: (df['DIQ050'] == 1).sum() / df['DIQ050'].notna().sum() * 100)
            plt.figure(figsize=(10,6))
            sns.lineplot(x=insulin.index, y=insulin.values, marker='o', color='red')
            plt.title('Insulin Use (%) by NHANES Cycle')
            plt.ylabel('Percent Using Insulin')
            plt.xlabel('NHANES Cycle')
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()

# --- Main Execution Block ---
if 'nhanes_classified' in globals() and not nhanes_classified.empty:
    print("📥 Loading merged DIQ data with age...")
    diq_combined = load_merged_diq_with_age()

    print("🔍 Generating summary report...")
    summary = summarize_medians_and_diagnoses(nhanes_classified, diq_combined)

    print("📈 Generating plots...")
    plot_trends(nhanes_classified, diq_combined)
else:
    print("❌ Error: 'nhanes_classified' not found. Run classification step first.")
```

### Fasting Glucose Analysis (t-test: gender, physical activity)

```python
from scipy.stats import ttest_ind
import seaborn as sns
import matplotlib.pyplot as plt

# Check that required variables are present
if 'nhanes_classified' in globals() and not nhanes_classified.empty:
    required_cols = ['LBXGLU', 'RIAGENDR', 'PAD680']  # LBXGLU = fasting glucose, PAD680 = physical activity status
    if all(col in nhanes_classified.columns for col in required_cols):

        # --- 1️⃣ Gender t-test ---
        print("\n🔍 Comparing Fasting Glucose by Gender (t-test)")

        data_gender = nhanes_classified[['LBXGLU', 'RIAGENDR']].dropna()
        data_gender = data_gender[data_gender['RIAGENDR'].isin([1, 2])]  # 1=Male, 2=Female

        males = data_gender[data_gender['RIAGENDR'] == 1]['LBXGLU']
        females = data_gender[data_gender['RIAGENDR'] == 2]['LBXGLU']

        t_stat_g, p_val_g = ttest_ind(males, females, equal_var=False)  # Welch's t-test

        print(f"\nGender - Welch's t-test Results:")
        print(f"  Mean (Males)   : {males.mean():.2f}")
        print(f"  Mean (Females) : {females.mean():.2f}")
        print(f"  t-statistic    : {t_stat_g:.4f}")
        print(f"  p-value        : {p_val_g:.4e}")

        # Visualization
        plt.figure(figsize=(8, 5))
        sns.boxplot(x='RIAGENDR', y='LBXGLU', data=data_gender)
        plt.xticks([0, 1], ['Male', 'Female'])
        plt.ylabel("Fasting Glucose (mg/dL)")
        plt.title("Fasting Glucose Levels by Gender")
        plt.grid(True, linestyle='--', alpha=0.5)
        plt.show()

        # --- 2️⃣ Physical Activity t-test ---
        print("\n🔍 Comparing Fasting Glucose by Physical Activity (t-test)")

        # Assume PAD680: 1 = Yes (physically active), 2 = No
        data_pa = nhanes_classified[['LBXGLU', 'PAD680']].dropna()
        data_pa = data_pa[data_pa['PAD680'].isin([1, 2])]

        active = data_pa[data_pa['PAD680'] == 1]['LBXGLU']
        inactive = data_pa[data_pa['PAD680'] == 2]['LBXGLU']

        t_stat_pa, p_val_pa = ttest_ind(active, inactive, equal_var=False)

        print(f"\nPhysical Activity - Welch's t-test Results:")
        print(f"  Mean (Active)   : {active.mean():.2f}")
        print(f"  Mean (Inactive) : {inactive.mean():.2f}")
        print(f"  t-statistic     : {t_stat_pa:.4f}")
        print(f"  p-value         : {p_val_pa:.4e}")

        # Visualization
        plt.figure(figsize=(8, 5))
        sns.boxplot(x='PAD680', y='LBXGLU', data=data_pa)
        plt.xticks([0, 1], ['Active', 'Inactive'])
        plt.ylabel("Fasting Glucose (mg/dL)")
        plt.title("Fasting Glucose Levels by Physical Activity")
        plt.grid(True, linestyle='--', alpha=0.5)
        plt.show()

    else:
        print(f"❌ Required columns {required_cols} not found.")
else:
    print("❌ Error: 'nhanes_classified' not found. Run earlier steps first.")
```
